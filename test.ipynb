```python
import pandas as pd
```

# Read the data

```python
users_col_names = ['ID', 'Type', 'Name', 'Reserved field', 'Description']
users = pd.read_csv('data/users.csv', sep='|', on_bad_lines='skip', names=users_col_names)
```

```python
subscriptions_col_names = ['FollowerID', 'FollowedID']
subs = pd.read_csv('data/subscriptions.csv', sep=',', on_bad_lines='skip', names=subscriptions_col_names)
```

```python
services_col_names = ['UserID', 'ServiceID', 'ServiceName', 'ServiceURL', 'UserNameOnService', 'UserURLOnService']
services = pd.read_csv('data/services.csv', sep='|', on_bad_lines='skip', names=services_col_names)
```

```python
likes_col_names = ['userID', 'PostID', 'Timestamp']
likes = pd.read_csv('data/likes.csv', sep='\t', on_bad_lines='skip', names=likes_col_names)
```

```python
following_col_names = ['FollowerID', 'FollowedID', 'Timestamp']
following = pd.read_csv('data/followingAugSept.csv', sep='\t', on_bad_lines='skip', names=following_col_names)
```

```python
comments_col_names = ['PostID', 'EntryID', 'PostedBy', 'SourceName', 'SourceURL', 'GeoX', 'GeoY', 'Timestamp', 'Text', 'NumImg', 'ImgURL', 'NumVid', 'VidURL']
comments = pd.read_csv('data/commentAugSept.csv', sep='\t', on_bad_lines='skip', names=comments_col_names)
```

```python
entries_col_names = ['PostID', 'PostedBy', 'SourceName', 'SourceURL', 'GeoX', 'GeoY', 'Timestamp', 'Text', 'NumImg', 'ImgURL', 'NumVid', 'VidURL']
entries1 = pd.read_csv('data/entries1.csv', sep='\t', on_bad_lines='skip', names=entries_col_names)
entries2 = pd.read_csv('data/entries2.csv', sep='\t', on_bad_lines='skip', names=entries_col_names)
entries3 = pd.read_csv('data/entries3.csv', sep='\t', on_bad_lines='skip', names=entries_col_names)
```

# Drop unrelevant columns

```python
users.drop('Reserved field', axis=1, inplace=True)
```

```python
comments.drop(['GeoX', 'GeoY', 'NumImg', 'NumVid', 'ImgURL', 'VidURL'], axis=1, inplace=True)
```

```python
entries1.drop(['GeoX', 'GeoY', 'NumImg', 'NumVid', 'ImgURL', 'VidURL'], axis=1, inplace=True)
```

```python
entries2.drop(['GeoX', 'GeoY', 'NumImg', 'NumVid', 'ImgURL', 'VidURL'], axis=1, inplace=True)
```

```python
entries3.drop(['GeoX', 'GeoY', 'NumImg', 'NumVid', 'ImgURL', 'VidURL'], axis=1, inplace=True)
```
